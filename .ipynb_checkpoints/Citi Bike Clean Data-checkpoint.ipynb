```python
# Import Dependencies 
import pandas as pd
import datetime
```

```python
# Add in CSV's 
jan_csv = "Resources/Jan-2020-data.csv"
feb_csv = "Resources/Feb-2020-data.csv"
mar_csv = "Resources/Mar-2020-data.csv"
apr_csv = "Resources/Apr-2020-data.csv"
may_csv = "Resources/May-2020-data.csv"
```

```python
# Create df's for each month and show them
jan_df = pd.read_csv(jan_csv)
feb_df = pd.read_csv(feb_csv)
mar_df = pd.read_csv(mar_csv)
apr_df = pd.read_csv(apr_csv)
may_df = pd.read_csv(may_csv)
```

```python
# concat all df's together 
all_months_df = pd.concat([jan_df, feb_df, mar_df, apr_df, may_df], axis=0)

# Display results
all_months_df
```

```python
# Alter gender column 
gender_count = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use map to alter gender values 
all_months_df['gender'] = all_months_df['gender'].map(gender_count)

# Display updated df
all_months_df
```

```python
# Calculate current year to count ages
current_year = datetime.datetime.now().year
current_year
```

```python
# Count ages based on current year
all_months_df['Age'] = current_year - all_months_df['birth year']

# Define age groups
age_groups = [(16,19), (20,29), (30,39), (40,49), (50,59), (60,100)]
```

```python
# Create a column for age groups
all_months_df['Age Group'] = pd.cut(all_months_df['Age'], bins=[start for start, end in age_groups] + [age_groups[-1][1] + 1], labels=['16-19', '20-29', '30-39', '40-49', '50-59', '60+'])
```

```python
print(all_months_df)
```

```python
# Extract the date and monthcomponent from the starttimecolumn
all_months_df['Date'] = pd.to_datetime(all_months_df['starttime']).dt.date
all_months_df['Month'] = pd.to_datetime(all_months_df['starttime']).dt.strftime('%b')
```

```python
# Group the age groups and gender to count occurrences 
age_gender_month = all_months_df.groupby(['Month','Age Group', 'gender']).size().reset_index(name='Count')
```

```python
#Display results
print(age_gender_month)
```

```python
print(all_months_df)
```

```python
# Find the total number of trips taken per month
trips_per_month = all_months_df.groupby('Month').size()
print(trips_per_month)
```

```python
total_trips = len(all_months_df)
print(total_trips)
```

```python
all_months_df['Total Trips'] = total_trips
```

```python
all_months_df['Trips per Month'] = all_months_df['Month'].map(trips_per_month)
```

```python
# Save the updated df to a csv file 
all_months_df.to_csv('Resources/new_data.csv', index=False)
```
